```python
import pandas as pd
import datetime as dt
# Daftar file Excel yang akan digabungkan
file_paths = ['1.xlsx','2.xlsx','3.xlsx','4.xlsx','5.xlsx','6.xlsx','7.xlsx',
              '8.xlsx','9.xlsx','10.xlsx','11.xlsx','12.xlsx','13.xlsx','14.xlsx',
              '15.xlsx','16.xlsx','17.xlsx','18.xlsx','19.xlsx','20.xlsx','21.xlsx',
              '22.xlsx','23.xlsx','24.xlsx','25.xlsx','26.xlsx','27.xlsx','28.xlsx',
              '29.xlsx','30.xlsx','31.xlsx','32.xlsx','33.xlsx','34.xlsx','35.xlsx',
              '36.xlsx','37.xlsx','38.xlsx','39.xlsx','40.xlsx','41.xlsx','42.xlsx',
              '43.xlsx']

# Inisialisasi list untuk menyimpan DataFrame dari setiap file Excel
dfs = []

# Baca setiap file Excel dan tambahkan DataFrame ke dalam list
for file_path in file_paths:
    df = pd.read_excel(file_path)
    dfs.append(df)

# Gabungkan DataFrame menggunakan pd.concat()
combined_df = pd.concat(dfs, ignore_index=True)

# Simpan DataFrame hasil penggabungan ke dalam file Excel
combined_df.to_excel('Raw Data.xlsx', index=False)
```

```python
# Baca file Excel ke dalam DataFrame
df = pd.read_excel('rawr.xlsx')

# Identifikasi baris yang memiliki nilai 'Permintaan Disetujui' di kolom 'Status Pembatalan/ Pengembalian'
rows_to_drop = df[df['Status Pembatalan/ Pengembalian'] == 'Permintaan Disetujui'].index

# Hapus nilai yang sesuai
df.drop(rows_to_drop, inplace=True)

# # Simpan DataFrame yang sudah dimodifikasi kembali ke file Excel
# df.to_excel('nama_file_hasil.xlsx', index=False)
```

```python
# Tentukan kolom yang ingin dihapus
kolom_yang_dihapus = [kolom for kolom in df.columns if kolom not in ['Waktu Pesanan Dibuat', 'Username (Pembeli)', 'Total Harga Produk']]

# Buat DataFrame baru dengan hanya menyertakan kolom yang diinginkan
df_baru = df.drop(columns=kolom_yang_dihapus)
```

```python
df_baru.head()
```

```python
df_baru = df_baru.rename(columns={'Waktu Pesanan Dibuat': 'date','Total Harga Produk':'paid_price','Username (Pembeli)':'username'})
```

```python
df_baru.info()
```

```python
# Mengonversi kolom 'Waktu Pesanan Dibuat' menjadi tipe data datetime
df_baru['date'] = pd.to_datetime(df_baru['date'])
df_baru['paid_price']=df_baru['paid_price'].astype(int)

# Mengambil hanya bagian tanggal (hari)
df_baru['date'] = df_baru['date'].dt.date
df_baru['date'] = pd.to_datetime(df_baru['date'])
```

```python
present = dt.datetime(2024,5,1)
rfm = cleaned_data.groupby('username').agg({
    'date': lambda date: (present - dt.datetime.combine(date.max(), dt.datetime.min.time())).days,
    'username': lambda num: len(num),
    'paid_price': lambda price: price.sum()
})
```

```python
rfm.columns=['recency','frequency', 'monetary']
```
